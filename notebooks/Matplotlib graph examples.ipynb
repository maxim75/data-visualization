```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Line

```python
plt.style.use("seaborn-whitegrid")

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
x2 = np.linspace(0,10,10)

plt.ylabel("Y Label")

plt.plot(x, np.sin(x)*20, c="red")
plt.plot(x2, x2**2, c="blue", marker=".", markersize=10)

ax.axvline(x = 2.5, color='orange', linestyle=":") # linestyle: "-", "--", ":"
```

# Scatter

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.scatter(x2, x2**2, c=["red"]*5+["blue"]*5, s=x2**3, alpha=0.5)
```

# Horizontal Bar

```python
x = np.linspace(0,10,11)

fig, ax = plt.subplots(nrows = 1, ncols = 1)

ax.set_title("Horizontal Bar")

rects = ax.barh(x, x,
    height=0.5,
    color=["red"] + ["blue"]*9,
    tick_label=[f"Value {z}" for z in x]
)

ax.set_xlim(0,11)
ax.set_ylim(0,11)
```

# Horizontal Bar stacked

```python
x = np.linspace(1,10,10)

fig, ax = plt.subplots(nrows = 1, ncols = 1)

ax.set_title("Bar", fontsize=20, pad=30)

# blue bars
ax.barh(x, x,
    left=.5,
    color=["blue"]*len(x),
    tick_label=[f"Value {z}" for z in x],
    label="blue",
    edgecolor = "none")

# red bars
rects = ax.barh(x, 
                x**.5,
    left=x,
    color= ["red"]*len(x),
    tick_label=[f"Value {z}" for z in x],
    label="red",
    edgecolor = "none")

ax.set_yticks(x, minor=False)

ax.legend(loc='upper right', bbox_to_anchor=(1.2, 1), frameon=False)

for spine in ax.spines:
    ax.spines[spine].set_visible(False)
    
for i, v in enumerate(x):
    ax.text(v + v**.5 + .2, i+.8, int(v), color='black', fontweight='normal')
    
ax.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off

ax.grid()
```

# Pie

```python
x = np.linspace(1,5,5)

sns.set()
pie, ax = plt.subplots(figsize=[10,6])
labels = [f"Value {int(z)}" for z in x]
colors = ["#999"]*(len(x)-2) + ["#f00"] + ["#0f0"]

ax.pie(x=x, autopct="%.1f%%", 
        labels=labels, 
        pctdistance=0.7, 
        colors=colors, 
        textprops={'fontsize': 15})

ax.set_title("Show Pie", fontsize=18)
```
